```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
```

```python
casas=pd.read_csv('realtor-data.zip.csv')
casas.info()
```

```python
casas.head()
```

```python
#exercício 1.1
casas_tipo=pd.DataFrame(casas.dtypes).reset_index()

casas_limpo=casas.dropna(subset=list(casas_tipo[casas_tipo.iloc[:,1]=='object']['index']))
```

```python
#exercício 1.2
casas_limpo.bed=casas_limpo.bed.fillna(0)
casas_limpo.bath=casas_limpo.bath.fillna(0)
```

```python
#exercício 1.3
casas_limpo.house_size.fillna(np.median(casas_limpo[~casas_limpo.house_size.isna()].house_size)).value_counts(dropna=False)
```

```python
#exercício 2
casas_limpo['data_prev_ano']=pd.to_datetime(casas_limpo.prev_sold_date,format='%Y-%m-%d', errors='coerce').dt.year
pd.pivot_table(casas_limpo, index='state', columns='data_prev_ano', values='price', aggfunc='sum')
```
