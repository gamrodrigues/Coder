# Matplotlip

Pyplot

```python
#importações padrões

import pandas as pd
import numpy as np
```

```python
#importação pyplot
import matplotlib.pyplot as plt
```

```python
#y=2x
x=list(range(11))
y=list(pd.Series(x)*2)
plt.plot(x,y)
```

```python
#y=-x²+2x-10
x=list(range(-10,13))
y=list(pd.Series(x)**2*(-1)+2*pd.Series(x)-10)
plt.plot(x,y)
```

```python
#orientada ao estado
plt.plot([1,2],[3,4])
```

```python
#orientada ao objeto
fig,ax = plt.subplots()
ax.plot([1,2],[3,4])
```

```python
##Tipos de gráficos
x=list(range(5))
y=list(pd.Series(x)**2+2)
z=['Alfa', 'Beta', 'Gamma', 'Epsilo','Lambda']
print('x: ',x,"\n",'y: ',y,"\n",'z: ',z)
```

```python
#linha
fig,ax = plt.subplots()
ax.plot(x,y)
```

```python
#barras ou colunas
fig,ax = plt.subplots()
ax.bar(z,y)
```

```python
#pizza ou setores
fig,ax = plt.subplots()
ax.pie(y, labels=z)
```

```python
#dispersão
fig,ax = plt.subplots()
ax.scatter(x,y)
```

```python
#dispersão
np.random.seed(18)
k=np.random.randint(0,10,100)
fig,ax = plt.subplots()
ax.hist(k)
```

```python
fig,ax = plt.subplots()
ax.hist(k, bins=5)
```

```python
#área
fig,ax = plt.subplots()
ax.fill_between(x,y)
```

```python
#Configurações
```

```python
fig,ax = plt.subplots()
#cor 
ax.plot(x,y, color='lightgreen')
#titulo
ax.set_title("Gráfico de linha X e Y")
#etiqueta dos eixos
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
#limites dos eixos
ax.set_ylim(0)
#cor
ax.set_xscale('log')
```

```python
fig,ax = plt.subplots()
#cor 
ax.plot(x,y, color='lightgreen')
#tipo de linha e marcadores
ax.plot(x,pd.Series(y)-5, color='magenta', ls='--', lw=0.5, marker='.', markersize=20)
#transparencia
ax.plot(x,[1]*len(x), alpha=0.1, color='orange')
#titulo
ax.set_title("Gráfico de linha X e Y")
#etiqueta dos eixos
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
#limites dos eixos
ax.set_xlim(-1)
ax.set_ylim(-5)
ax.grid(True)
```

```python
#subplot
fig,axs = plt.subplots(2,2, figsize=(10,8))
axs[0,0].plot(x,y)
axs[0,0].set_title('Grafico linha X²+2')
axs[1,0].bar(z,y)
axs[1,0].set_title('Grafico de barras')
axs[0,1].hist(k, bins=4)
axs[0,1].set_title('Histrograma')
axs[1,1].pie(y, labels=z)
axs[1,1].set_title('Grafico de Pizza')
```
