```python
import pandas as pd
```

```python
df=pd.read_csv('realtor-data.zip.csv')
df.info()
```

```python
df.head()
```

```python
df_filtro=df[df.status=='for_sale'].copy()
df_filtro.info()
```

```python

df_filtro.sort_values('house_size', ascending=False).drop_duplicates(subset=['status', 'city', 'state', 'price'], inplace=True)
df_filtro.info()
```

```python
df_filtro.reset_index(drop=True, inplace=True)
df_filtro.state.value_counts()
```

```python
df_filtro.dropna(subset='state', inplace=True)
```

```python
amostra_df=pd.DataFrame()

amostra_df=pd.concat([amostra_df,df[df.state=='Massachusetts'].sample(n=250, random_state=25)])
amostra_df=pd.concat([amostra_df,df[df.state=='Puerto Rico'].sample(n=250, random_state=25)])
amostra_df=pd.concat([amostra_df,df[df.state=='Connecticut'].sample(n=250, random_state=25)])
amostra_df=pd.concat([amostra_df,df[(df.state=='Rhode Island')|(df.state=='New York')].sample(n=250, random_state=25)])
amostra_df
```

```python
amostra_df_filtro=pd.DataFrame()

amostra_df_filtro=pd.concat([amostra_df_filtro,df_filtro[df_filtro.state=='Massachusetts'].sample(n=250, random_state=25)])
amostra_df_filtro=pd.concat([amostra_df_filtro,df_filtro[df_filtro.state=='Puerto Rico'].sample(n=250, random_state=25)])
amostra_df_filtro=pd.concat([amostra_df_filtro,df_filtro[df_filtro.state=='Connecticut'].sample(n=250, random_state=25)])
amostra_df_filtro=pd.concat([amostra_df_filtro,df_filtro[(df_filtro.state=='Rhode Island')|(df_filtro.state=='New York')].sample(n=250, random_state=25)])
amostra_df_filtro
```
