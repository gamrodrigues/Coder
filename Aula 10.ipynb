```python
#Bibliotecas
import pandas as pd
import numpy as np
```

```python
#Base de dados
casas= pd.read_csv('Dataset - Preço de Casas.csv', sep=' ', encoding='latin1', decimal=',')
casas
```

```python
casas.info()
```

```python
#funções de agregação

print('Contagem:',casas['Preço($)'].count())
print('Soma:',casas['Preço($)'].sum())
print('Média:',casas['Preço($)'].mean())
print('Mínimo:',casas['Preço($)'].min())
print('Máximo:',casas['Preço($)'].max())
print('Mediana:',casas['Preço($)'].median())
print('Quantil 0%:',casas['Preço($)'].quantile(0))
print('Quantil 25%:',casas['Preço($)'].quantile(0.25))
print('Quantil ?%:',casas['Preço($)'].quantile())
print('Quantil 50%:',casas['Preço($)'].quantile(0.5))
print('Quantil 75%:',casas['Preço($)'].quantile(0.75))
print('Quantil 100%:',casas['Preço($)'].quantile(1))
print('Desvio-padrão:',casas['Preço($)'].std())
print('Variância:',casas['Preço($)'].var())
```

```python
casas.describe(include='all')
```

```python
casas_str=casas.copy()
casas_str['Quartos']=casas_str['Quartos'].astype(str)
casas_str['Banheiros']=casas_str['Banheiros'].astype(str)
casas_str.describe(include='all')
```

```python
#group by
#df.groupby('coluna string para agregar')[['variáveis quantitativas']].'função de agregação'
casas.groupby('Vizinhança')[['Quartos']].mean()
```

```python
casas.groupby('Vizinhança')[['Quartos','Banheiros']].mean()
```

```python
casas.groupby('Vizinhança')[['Preço($)']].sum()#/1000000
```

```python
casas.groupby('Vizinhança')[['AnoConstrução']].min()
```

```python
casas.groupby('Vizinhança')[['AnoConstrução']].max()
```

```python
casas_str.groupby('Quartos')[['Preço($)']].max()
```

```python
casas_str.groupby('Banheiros')[['Preço($)']].max()
```

```python
#group by agg

casas.groupby('Vizinhança').agg(['min', 'max'])
```

```python
#Ordenação
casas.sort_values('AnoConstrução')
```

```python
casas.sort_values('AnoConstrução', ascending=False)
```

```python
sme1=pd.read_excel('Dataset - Saude Mental Estudantes.xlsx', sheet_name='Dataset - Saude Mental 1')
sme1.head()
```

```python
sme2=pd.read_excel('Dataset - Saude Mental Estudantes.xlsx', sheet_name='Dataset - Saude Mental 2')
sme2.head(15)
```

```python
sme3=pd.read_excel('Dataset - Saude Mental Estudantes.xlsx', sheet_name='Dataset - Saude Mental 3')
sme3.tail()
```

```python
print(pd.concat([sme1,sme2]))
print(pd.concat([sme1,sme2]).index.value_counts().sort_index())
```

```python
sme_full=pd.concat([sme1,sme2], ignore_index=True)
print(sme_full)
print(sme_full.index.value_counts().sort_index())
```

```python
pd.concat([sme_full,sme3])
```

```python
sme_full=pd.concat([sme_full,sme3], axis=1)
sme_full
```

```python
queijos=pd.read_csv('Dataset - cheeses.csv')
queijos.tail()
```

```python
queijos.merge(queijos_detalhes, how='right', on='url')
```
