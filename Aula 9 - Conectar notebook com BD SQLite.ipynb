```python
import sqlite3

# Criar uma conexão e um cursor
# Salve o arquivo coderhouse.db na sessão
#Atualize o nome do arquivo
conn = sqlite3.connect('coderhouse.db')
cursor = conn.cursor()

# Fazer uma leitura de exemplo
cursor.execute('''
SELECT * FROM fastfood f;
''')
leitura_exemplo = cursor.fetchall()
print(leitura_exemplo)

# Salvar as alterações e fechar a conexão
conn.commit()
conn.close()
```
