```python
import re
import requests
from bs4 import BeautifulSoup
import txt
```

```python
#Exercício 1
while True:
    email=input('Digite um email')
    if not(re.fullmatch( r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,7}\b', email)): 
        print(email,'\t-->\tE-mail invalido, tente novamente.')
    else:
        print(email,'\t-->\tE-mail válido')
        break
```
